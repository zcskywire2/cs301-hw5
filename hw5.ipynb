```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
try:
    df = pd.read_csv("student_class.csv")
except Exception as e:
    print (e.__doc__)
    print (e)
    print("Error during reading, please try again")
mid = df.groupby(['Majors'])['Midterm'].mean().tolist()
fin = df.groupby(['Majors'])['Final'].mean().tolist()
avg = df.groupby(['Majors'])['Final Average'].mean().tolist()
ind = np.arange(5) 
width = 0.3       
plt.bar(ind, mid, width, label='Midtern mean')
plt.bar(ind + width, fin, width,label='Final mean')
plt.bar(ind + (width*2), avg, width,label='Final average mean')

plt.ylabel('Mean Grades')
plt.title('Mean Grades by major')

plt.xticks(ind + width, ('BNFO', 'CS', 'IS', 'IT', 'MATH'))
plt.legend(loc='best')
plt.show()
print()
print('Over all it looks like the BNFO major is doing the best as over all they have the highest grades of all the major, espically at the end of the classes')


x = df['Final Average'].tolist()
x.extend(df['Final'].tolist())
x.extend(df['Final Average'].tolist())
num_bins = 10
n, bins, patches = plt.hist(x, num_bins, facecolor='blue', alpha=0.5)
plt.ylabel('Frequency')
plt.xlabel('Grade')
plt.title('Distrubtion of all grades')
plt.show()
print()
print("The distrubtion of the graph is reltivly normal distrubtion but there are some odd spikes near the middle.")
print('I cant think at the moment of anything id change on the graph')


g1 = (df['Midterm'].tolist(),df['Final Average'].tolist())
g2 = (df['Final'].tolist(),df['Final Average'].tolist())
N = len(g1)+len(g2)
data = (g1, g2)
colors = ("red", "blue")
groups = ("Midterm", "Final")

# Create plot
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, facecolor="1.0")

for data, color, group in zip(data, colors, groups):
    x, y = data
    ax.scatter(x, y, alpha=0.8, c=color, edgecolors='none', s=30, label=group)

plt.ylabel('Final Average')
plt.xlabel('Exam Score')
plt.title('Exam vs Final Average')
plt.legend(loc=2)
plt.show()
print()
print('This plot shows that there is a positive corellation between the grades on the midterm or final, and the final average of the student. This can be seen in the plot by the upward trend as you move to the right on the graph. ')
```
